### 向量化实现SimpleLinearaRegression

```python
from LinearRegress.SimpleLinearRegressionB import SimpleLinearRegression
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.,2.,3.,4.,5.])
y = np.array([1.,3.,2.,3.,5.])
```

```python
reg2 = SimpleLinearRegression()
reg2.fit(x,y)
```

```python
reg2.a_
```

```python
reg2.b_
```

```python
y_hat = reg2.predict(x)
plt.scatter(x, y)
plt.plot(x, y_hat, color='r')

plt.axis([0,6,0,6])
```

### 向量化实现的性能测试

```python
m = 1000000
big_x = np.random.random(size=m)
big_y = big_x * 0.2 +3.0 +np.random.normal(size=m)
```

```python
%time reg2.fit(big_x,big_y)
```
