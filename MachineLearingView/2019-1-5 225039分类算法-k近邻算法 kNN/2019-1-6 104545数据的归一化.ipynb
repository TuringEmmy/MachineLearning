对于肿瘤案例中，样本空间的距离发现时间所主导

解决方案：将所有的数据映射到同一尺度

最值归一化：把所有的数据映射到0-1之间

### 最值归一化normalization



适用于分布有明县边界的情况；受outlier影响较大

### 均值方差归一化standzrdization

数据分布没有明显的边界；有可能存在极端的数据值

均值方差归一化：把所有的数据归一到俱进之为0方差为1 的分布中


```python
import numpy as np
```

### 最值归一化Normalization

```python
x = np.random.randint(0,100,size=100)
x
```

```python
(x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
X = np.random.randint(0,100,(50,2))
X[:3,:]
```

因为归一化是小数所有的把X的类型转换成float类型

```python
X = np.array(X, dtype=float)
X[:3,:]
```

```python
X[:,0 ] = (X[:,0] - np.min(X[:,0])) / ((np.max(X[:,0]) - np.min(X[:,0])))
X[:,1 ] = (X[:,1] - np.min(X[:,1]) )/ ((np.max(X[:,1]) - np.min(X[:,1])))
```

```python
X[:3,:]
```

对于样本特征多的时候需要for循环

```python
from matplotlib import pyplot  as plt
```

```python
plt.scatter(X[:,0],X[:,1],color='r')
```

```python
np.mean(X[:,0])
```

方差

```python
np.std(X[:,0])
```

### 均值方差归一化Standardization

```python
X2 = np.random.randint(0,50,(50,2))
X2 = np.array(X2, dtype=float)
```

```python
X2[:,0] = (X2[:,0] - np.mean(X2[:,0])) / (np.std(X2[:,0]))
X2[:,1] = (X2[:,1] - np.mean(X2[:,1])) / (np.std(X2[:,1]))
```

```python
plt.scatter(X2[:,0],X2[:,1])
```

```python
np.mean(X2[:,0])
```

```python
np.std(X2[:,0])
```

总结:最值归一化对于个别异常点无法处理,而方差归一化可以处理这种情况
