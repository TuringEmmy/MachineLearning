```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

画双图

```python
plt.plot(x,siny)
plt.plot(x,cosy)
```

改颜色

```python
plt.plot(x,siny,color='red')
plt.plot(x,cosy)
```

线条样式

```python
plt.plot(x,siny,color='red',linestyle='--')
plt.plot(x,cosy)
```

改x轴范围

```python
plt.plot(x,siny,color='red',linestyle='--')
plt.plot(x,cosy)
plt.xlim(-5,15)
```

一个函数调节

```python
plt.plot(x,siny,color='red',linestyle='--')
plt.plot(x,cosy)
plt.axis([-1,11,-2,2])
```

改轴的显示

```python
plt.plot(x,siny,color='red',linestyle='--')
plt.plot(x,cosy)
plt.axis([-1,11,-2,2])
plt.xlabel(" Turing")
plt.ylabel(" Emmy")
```

添加图例

```python
plt.plot(x,siny,color='red',linestyle='--', label='sin(x)')
plt.plot(x,cosy,label="cos(x)")
plt.axis([-1,11,-2,2])
plt.xlabel(" Turing")
plt.ylabel(" Emmy")
plt.legend()
```

图添加标题

```python
plt.plot(x,siny,color='red',linestyle='--', label='sin(x)')
plt.plot(x,cosy,label="cos(x)")
plt.axis([-1,11,-2,2])
plt.xlabel(" Turing")
plt.ylabel(" Emmy")
plt.title("This is a figure in chrom")
plt.legend()
```
