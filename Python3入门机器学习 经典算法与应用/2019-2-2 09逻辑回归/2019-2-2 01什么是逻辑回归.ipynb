### 什么是逻辑回归

逻辑回归：解决分类问题

将样本的特征和样本发生的概率联系起来，概率是一个数

逻辑回归既可以看做是回归算法，也可以看做是分类算法

通常作为分类算法用，只可以解决二分类问题

### sigmoid函数



```python
import numpy as np
import matplotlib.pyplot as plt
```

编写sigmoid函数

```python
def sigmoid(t):
    return 1 / (1 + np.exp(-t))
```

准备参数

```python
x = np.linspace(-10, 10,500)
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.grid()
```

分析上面函数的性质
