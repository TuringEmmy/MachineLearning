### sigmoid



```python
import numpy as np
import matplotlib.pyplot as plt
```

sigmoid

```python
def sigmoid(t):
    return 1 / (1 + np.exp(-t))
```

```python
x = np.linspace(-10, 10,500)
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.grid()
```
