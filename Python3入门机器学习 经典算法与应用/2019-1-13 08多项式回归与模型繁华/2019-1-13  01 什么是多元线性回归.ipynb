像这种。x^2和x相当于两个特征值的情况

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
```

```python
y = 0.5 * x**2 + x* 2 + np.random.normal(0,1, size=100)
```

```python
plt.scatter(X,y)
```

```python
X.shape
```

```python
y.shape
```

使用线性来拟合

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(X, y_predict,color='r')
```

上面的直线不能很好的表达散点的关系，所以决绝方案，添加一个特征

```python
(X**2).shape
```

```python
X2 = np.hstack([X, X**2])
```

```python
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y2_predict = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y2_predict[np.argsort(x)],color='r')
```

```python
lin_reg2.coef_
```

这个系数就是方程前面的系数

```python
lin_reg2.intercept_
```
