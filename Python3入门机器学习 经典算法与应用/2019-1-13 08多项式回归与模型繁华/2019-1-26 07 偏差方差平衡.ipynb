### 偏差方差权衡

Bias Variance Trade off

### 偏差和方差

问题可以理解成这个靶子，中心，而我么根据数据进行拟合，仅为预测这个模型，拟合的 模型就是这些打出去的点

### 模型误差

模型误差 = 偏差(Bias) + 方差（Variance） + 不可避免的误差

导致偏差的主要原因;

对问题本身的假设不正确!

如：非线性数据使用线性回归

过拟合overfitting就会出现方差比较大的情况

有一些算法天生是高方差的算法。如KNN. 非参数学习通常都是高方差算法。因为不对数据进行假设

有一些算法天生发稿偏差算法。如线性回归  参数学习通常都是高偏差算法。 因为对数据有极强的假设

参数学习通常都是高偏差算法。因为对数据具有极强的假设

大多数算法具有相应的参数，可以调整偏差和方差

如knn中的K

如线性回归中使用多项式回归

> 偏差和方差通常是矛盾的
> 降低偏差，会提高方差
> 降低方差，会提高偏差

### 方差

机器学习的主要挑战，来自于方差！

解决高方差的通常手段：

1. 降低模型复杂度

2. 减少数据维度；降噪

3. 增加样本数

4. 使用验证集

5. 模型的正则化
