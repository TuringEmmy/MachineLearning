常见案例[房价预测]

寻找一条直线,最大程度的"拟合"样本特征和样本输入之间的关系,横轴是特征,中州是预测,需要占用一个坐标轴

样本特征只有一个,称为简单线性回归

如果直接做差,多值相加可能会是0,舍弃


直接添加绝对值,也不可以,因为我们需要处处可导


最终结果,且处处可导


最终为



损失函数(loss functio)     效用函数(utility function)

通过分析问题,确定问题的瞬时函数或者小勇函数

通过最后花损失函数或者效用函数,获得机器学习的模型

> 所有参数学习算法都是这样的套路

线性回归  多项式回归  逻辑回归  SVM  神经网络  ...

最优化原理  凸优化.
