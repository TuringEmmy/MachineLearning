创建更多的子模型！继承更多的子模型的意见，子模型之间不能一致！子模型之间要有差异性

如何创建差异性？

每个子模型只看样本数据的一部分

**eg** 一共呢有500个样本数据，每个子模型只看100个样本数据

每一个子模型不需要太高的准确率

如果每个子模型只有51%的准确率, 如果我们只有一个子模型，整体准确率51%，如果我们有三个模型，整体的准确率


结果是51.5%

如果我们只有500个子模型，整体的准确率


结果为65.6%

当准确率为60%的时候，有500的子模型，准确率可以达到99.9999%

在算法比赛当中，集成学习的使用率非常的高

### Bagging 和 Pasting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

X, y = datasets.make_moons(n_samples=500, noise=0.3, random_state=42)
```

```python
plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=66)
```

### 使用Bagging

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bagging_clf = BaggingClassifier(DecisionTreeClassifier(),
                               n_estimators=500,max_samples=100,
                               bootstrap=True)
```

```python
bagging_clf.fit(X_train, y_train)
bagging_clf.score(X_test, y_test)
```

增加样本数

```python
bagging_clf2 = BaggingClassifier(DecisionTreeClassifier(),
                               n_estimators=5000,max_samples=100,
                               bootstrap=True)
```

```python
bagging_clf2.fit(X_train, y_train)
bagging_clf2.score(X_test, y_test)
```
