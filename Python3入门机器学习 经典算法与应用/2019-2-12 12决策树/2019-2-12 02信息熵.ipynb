### 信息熵

熵在信息论中代表随机变量不确定度的度量 

熵越大，数据的不确定性越高

具体数字计算


再极端的观察数据


## 二分类进行研究


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log(p)- (1-p)*np.log(1-p)
```

```python
x = np.linspace(0.01,0.99,200)
```

```python
plt.plot(x, entropy(x))
```

无论x趋于0或者趋于1，信息熵都变低了

如果是三类，就是一个立体的图形了

大纲多个类的概率是相同的时候，信息熵最高，也就是最不确定的时候
